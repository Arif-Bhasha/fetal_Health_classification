### Model selection and Building
set up features(x) and target(y)



```python
fv=data[data.columns[:-1]]
c_l=data[[data.columns[-1]]]
```

```python
c_l.head()
```

```python
c_l.value_counts()
```

spliting train and test sets

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(fv,c_l,train_size=0.8,stratify=c_l,random_state=100)
```

```python
y_test.value_counts()
```

```python
y_train.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
```

```python
x_train=pd.DataFrame(scale.fit_transform(x_train),columns=x_train.columns,index=x_train.index)
x_test=pd.DataFrame(scale.transform(x_test),columns=x_test.columns,index=x_test.index)
```

```python
from imblearn.over_sampling import ADASYN
```

```python
ad=ADASYN(sampling_strategy={1:1323,2:1323,3:1323})
```

```python
x_train,y_train=ad.fit_resample(x_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings("ignore")
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector
knn=KNeighborsClassifier()
ffs=SequentialFeatureSelector(knn,k_features=6,forward=True,n_jobs=-1)
ffs.fit(x_train,y_train)
features=list(ffs.k_feature_names_)
features
```

```python
x_train=x_train[features]
x_test=x_test[features]
x_test
```

```python
for y in range(1,20,2):
    l=cross_val_score(KNeighborsClassifier(n_neighbors=y),x_train,y_train,cv=5)
    print("for this {} == {}".format(y,np.mean(l)))
```

from the above cross_val_score we can observe that best "k" value is "3"

```python
knn=KNeighborsClassifier(n_neighbors=3)
kmodel=knn.fit(x_train,y_train)
knn_y_predict=kmodel.predict(x_test)
```

```python
tuned_parameters={'penalty':['l1','l2',],'C':[10**-4, 10**-2, 10**2, 10**4]}
```

```python
from sklearn.metrics import accuracy_score
lmodel=GridSearchCV(LogisticRegression(),tuned_parameters,scoring="accuracy",cv=5)
lgr=lmodel.fit(x_train,y_train)
print("tuned_parameters",lmodel.best_params_)
pr=lmodel.predict(x_test)
accuracy_score(y_test,pr)
```

```python
logreg=LogisticRegression(C=100,penalty='l2')
logreg.fit(x_train,y_train)
logreg_y_predict=logreg.predict(x_test)
```

```python
logreg.coef_
```

```python
gnb=GaussianNB()
```

```python
gmodel=gnb.fit(x_train,y_train)
```

```python
gmodel_predicted=gmodel.predict(x_test)
```

```python
print("knn_accuracy: ",accuracy_score(y_test,knn_y_predict))
print("logreg_accuracy: ",accuracy_score(y_test,logreg_y_predict))
print("gmodel_accuracy: ",accuracy_score(y_test,gmodel_predicted))
```
